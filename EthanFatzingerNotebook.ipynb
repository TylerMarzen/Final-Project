```python
import pandas as pd
import numpy as np
import geopandas
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("simplified_playing_fields_data.csv")
data
```

```python
plt.figure(figsize=(18,6))
plt.bar(x = data['Neighborhood'],
    height = data['Fields'])
plt.xticks(rotation = 90)
```

```python
neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
# neighborhoods.plot()
playingfield_map = neighborhoods.merge(data, how="left", left_on="hood", right_on="Neighborhood")
map = playingfield_map.plot(column='Neighborhood', # set the data to be used for coloring
               cmap='Greens',              # choose a color palette
               edgecolor="grey",        # outline the districts in white
               legend=True,              # show the legend
               legend_kwds={'label': "Number of Playingfields"}, # label the legend
               figsize=(15, 10),         # set the size
               missing_kwds={"color": "lightgrey"} # set disctricts with no data to gray
                   )

map.set_title("Number of Playing Fields in the Neighborhoods of Pittsburgh", fontsize = 15)
```
